# Experimenting with Characters

First we need a character set. I thought about trying to extract this from the definition file of a bitmap font, but that looked hard. A simpler thing would be just to draw some characters using an image library. Actually, we could just do that in matplotlib. Let's get some useful libraries.

```python
%matplotlib inline
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

The plan is to create a image containing each character which we can then use to extract properties like an equivalent brightness and orientation. To do this, we need to work out how big the matplotlib characters are.

Font size is specified in pt, which is roughly $\frac{1}{72}$th of an inch. We can get a more accurate value like this:

```python
# Specify a big font size
fontsize = 300

# Fiddle with this constant until the top of the ascender just hits the top of the axis
scale = 73.5

# Make a figure exactly 1 character high
figheight = float(fontsize)/scale
figwidth = float(fontsize)/scale

# Draw it
fig = plt.figure(figsize=(figwidth,figheight))
ax = fig.add_subplot(1,1,1)
ax.text(0, 0, "h", family="monospace", fontsize=fontsize)
fig.show()
```

If we use monospace characters then there will also be a constant width. We can do this in a similar manner:

```python
# Specify a big font size
fontsize = 300

# Font paramters
scale = 73.5

# Fiddle with this consant until the 7th . misses the figure
aspect = 0.793

# Make a figure large enough for exactly 6 characters
figheight = float(fontsize)/scale
figwidth = 6*float(fontsize)*aspect/scale

# Draw it
fig = plt.figure(figsize=(figwidth,figheight))
ax = fig.add_subplot(1,1,1)
ax.text(0, 0, ".......", family="monospace", fontsize=fontsize)
fig.show()

# Double check with w, because it's the widest character
fig = plt.figure(figsize=(figwidth,figheight))
ax = fig.add_subplot(1,1,1)
ax.text(0, 0, "wwwwww", family="monospace", fontsize=fontsize)
fig.show()
```

There's also the issue that our descending letters drop below the baseline. So we need to measure by how much

```python
# Specify a big font size
fontsize = 300

# Font paramters
scale = 73.5
aspect = 0.793

# Fiddle with this consant until the 7th . misses the figure
descent = 0.28

# Make a figure large enough for exactly 6 characters
figheight = fontsize*(1+descent)/scale
figwidth = 6*float(fontsize)*aspect/scale

# Draw it
fig = plt.figure(figsize=(figwidth,figheight))
ax = fig.add_subplot(1,1,1)
ax.text(0, descent/(1+descent), "hgbqfp", family="monospace", fontsize=fontsize)
fig.show()
```

Great. This is all the measurements we need to create a figure which snugly fits one character.
