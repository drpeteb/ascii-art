I don't really understand why the result is smaller... but no matter. Now we can loop through an entire set and make a little black and white image for each one.

```python
# Specify a font size
fontsize = 10

# Font paramters
scale = 95
aspect = 0.793
descent = 0.29
border = 0.1

# Characters to get
char_set = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ1234567890,./<>?;'#:@~[]{}-=_+)(*&^%$!\"\\`|"
# Output
char_dict = dict()

# Make a figure
figheight = (1+2*border)*fontsize*(1+descent)/scale
figwidth =  (1+2*border)*fontsize*aspect/scale
fig = plt.figure(figsize=(figwidth,figheight))
ax = fig.add_axes((0,0,1,1)) 

for cc in char_set:

    # Print the letter
    plt.cla()
    ax.text(0, descent/(1+descent), cc, family="monospace", fontsize=fontsize)

    # Remove the axes
    ax.set_frame_on(False)
    ax.set_xticks([])
    ax.set_yticks([])
    plt.axis('off')
    
    # Renderer
    fig.canvas.draw()

    # Convert to a numpy array
    w,h = fig.canvas.get_width_height()
    arr = np.fromstring( fig.canvas.tostring_rgb(), dtype=np.uint8 )
    arr.shape = (h,w,3)
    bwarr = np.mean(arr,axis=2)
    
    # Store it in the dictionary
    char_dict[cc] = bwarr
```

```python
def image_patch_orientation(im):
    """
    Calculate the orientation (direction and strength) of an image patch stored as a 2D numpy array.
    """
    
    # Get image size
    h,w = im.shape
    
    # Create arrays for the orientation and strength
    grad = np.zeros((h,w,2))
    orient = np.zeros((h,w))
    strength = np.zeros((h,w))
    
    # Loop through pixels
    for rr in range(h):
        for cc in range(w):
            
            # Gradient for this pixel
            pixel_grad = np.zeros(2)
            
            # First row order difference
            if cc==0:
                pixel_grad[0] = float(im[rr,cc+1]-im[rr,cc])
            elif cc==w-1:
                pixel_grad[0] = float(im[rr,cc]-im[rr,cc-1])
            else:
                pixel_grad[0] = float(im[rr,cc+1]-im[rr,cc-1])/2
                
            # First column order difference
            if rr==0:
                pixel_grad[1] = float(im[rr+1,cc]-im[rr,cc])
            elif rr==h-1:
                pixel_grad[1] = float(im[rr,cc]-im[rr-1,cc])
            else:
                pixel_grad[1] = float(im[rr+1,cc]-im[rr-1,cc])/2
            
            # Pixel orientation
            orient = np.arctan2(pixel_grad[1],pixel_grad[0])
            
            # Resolve direction ambiguity
            if orient<0:
                pixel_grad = - pixel_grad
            
            # Store
            grad[rr,cc,:] = pixel_grad
    
    # Mean gradient
    mean_grad = np.mean(grad,axis=(0,1))
    print(mean_grad)
    
    # Orientation
    orient = np.arctan(mean_grad[1]/mean_grad[0])
            
    return orient

# Test it out
orient = image_patch_orientation(char_dict["-"])
orient
            
```

```python
def image_patch_daisy(im):
    """
    Make a feature vector for an image patch using DAISY descriptors
    """
    
    fv = skimfeat.daisy(im, radius=3, rings=1)
    
    return np.ndarray.flatten(fv)

test_char = char_dict["A"]
print(test_char.shape)
fv = image_patch_daisy(test_char)
print(fv)
print(test_char)
```

```python
# Load an image
test_image = np.asarray(Image.open('Lenna.png'))
test_image = np.mean(test_image, axis=2)
plt.imshow(test_image)

# Block size
bh,bw = char_dict["a"].shape
```

```python
def asciiise(im,char_set,char_intensity,char_daisy,daisy_scale,bh,bw):
    """
    Create ascii art string for an image
    """
    
    # number of blocks
    h,w = im.shape
    nc = int(np.floor(w/bw))
    nr = int(np.floor(h/bh))
    
    # Empty array
    asc = []
    
    # Loop through blocks
    for rr in range(nr):
        for cc in range(nc):
            
            # Patch
            patch = im[rr*bh:((rr+1)*bh-1), cc*bw:((cc+1)*bw-1)]
            
            # Get feature vectors
            fv_daisy = image_patch_daisy(patch)
            fv_intensity = image_patch_intensity(patch)
            
            # Find minimum distance
            dist1 = [np.linalg.norm(fv_intensity-csfv) for csfv in char_intensity]
            dist2 = [daisy_scale*np.linalg.norm(fv_daisy-csfv) for csfv in char_daisy]
            
            #print(min(dist1))
            #print(min(dist2))
            
            from operator import add
            dist = map(add, dist1, dist2)
            idx = dist.index(min(dist))
            #print(idx)
            
            # Get characted
            asc.append(char_set[idx])
            
        asc.append("\n")
            
    print(''.join(asc))
    
daisy_scale = 1
asciiise(test_image, char_set, char_intensity, char_daisy, daisy_scale, bh, bw)
```
