# Introduction

I've often wondered how ascii art is made. You know... <a href="http://en.wikipedia.org/wiki/ASCII_art">this stuff</a>. It's one of those things that seemed a bit magic. So one day I decided to find out how it works, and maybe try to make some of my own. This is the result.

Obviously the first stop was Wikipedia, then <a href="http://stackoverflow.com/questions/394882/how-do-ascii-art-image-conversion-algorithms-work">stackoverflow</a>, which led me to a nice little introductory page <a href="http://www.jave.de/image2ascii/algorithms.html">here</a>, as well as to an <a href="http://aa-project.sourceforge.net/aalib/">extensive ascii art library</a> and a <a href="http://www.chris.com/ascii/">repository of more ascii art than I could ever need</a>.

The basic strategy is to consider the brightness of each pixel or pixel patch (in black and white), and then choose an ascii character with a similar brightness to represent it. Many schemes seem to do this a bit heuristically. Choose a set of characters with some sort of gradient, say " .,o0@", then set thresholds at brightness levels to choose which one to use. I think we can probably do better, by using the whole ascii set and doing a nearest neighbour match. Also, I think we can probably use additional information, orientation of the character and such.

Time to get started. First let's grab some useful libraries which we might want:

# Character Brightness

The first, and most essential, feature to look at is simply the brightness of each character. Let's make a function to get that:

```python
def image_patch_brightness(im):
    """
    Calculate the brightness of an image patch stored as a 2D numpy array.
    """
    
    return np.mean(im)

# Test it
image_patch_brightness(char_dict["a"])
```

That was easy. Let's calculate the brightness for each one in the dictionary:

```python
brightness_list = []
for cc in char_set:
    
    # Get the feature
    brightness = image_patch_brightness(char_dict[cc])
    
    # Add to list
    brightness_list.append(brightness)

# Plot a histogram of brightnesses
plt.figure()
plt.hist(brightness_list, 20)

# What are the maximum and minimum values?
print(np.max(brightness_list))
print(np.min(brightness_list))
```

They're not very well spread out. In particular, there's nothing below about 170. We'll need to bear this in mind and rescale if necessary.

# Drawing Ascii Art

We're ready to give it a go. Here's the main function to ascii-ise an image:

```python
def asciiise(image, char_set, brightness_list, block_height, aspect_ratio):
    """
    Create ascii art string for an image
    """
    
    # Block size
    bh = int(block_height)
    bw = int(np.floor(block_height*aspect_ratio))
    
    # Number of blocks
    h,w = image.shape
    nc = int(np.floor(w/bw))
    nr = int(np.floor(h/bh))
    
    # Brightness grid
    brightness_grid = np.zeros((nr,nc))
    
    # Empty list to build our picture in
    asc = []
    
    # Loop through blocks
    for rr in range(nr):
        for cc in range(nc):
            
            # Grab an image patch
            patch = image[rr*bh:((rr+1)*bh-1), cc*bw:((cc+1)*bw-1)]
            
            # Get it's brightness
            brightness_grid[rr,cc] = image_patch_brightness(patch)
    
    # Rescale
    brightness_grid = (brightness_grid - np.min(brightness_grid) ) / (np.max(brightness_grid)-np.min(brightness_grid))
    brightness_list = (brightness_list - np.min(brightness_list) ) / (np.max(brightness_list)-np.min(brightness_list))
    
    # Loop through blocks
    for rr in range(nr):
        for cc in range(nc):
            
            # Block rescaled brightness
            bright = brightness_grid[rr,cc]
            
            # Find minimum distance
            dist = [np.abs(bright-cb) for cb in brightness_list]
            idx = dist.index(min(dist))
            
            # Add character to the picture
            asc.append(char_set[idx])
            
        # New line
        asc.append("\n")
    
    # Print it
    print(''.join(asc))
```

This is pretty inefficient, but it'll do for now. Let's try it out. We'll need a test image. Hmmmm, what to use? Oh yeah.

The question now is how to make it better. The obvious answer is to make the characters smaller, but that's cheating really. More desirable would be to get rid of the background. If we could set that all to white, then it would become much clearer, but that's a really tough thing to do without manual intervention.

What we could do is try to make more use of the information in the image patches. At the moment, we're just matching the brightness, but there's more there. How about using gradient information?

# Image Features

Scikit-image has some built-in feature descriptors which we could use to match characters to image patches. No SIFT, but it does have something called DAISY which is decribed as "SIFT-like", some sort of histogram of orientations thing. Let's try that:

```python
def image_patch_feature(im):
    """
    Make a feature vector for an image patch using DAISY descriptors
    """
    
    # Set the radius of the features so that there's exactly one per character/patch
    radius = int(im.shape[1]/2)
    
    # Make a feature
    fv = skimfeat.daisy(im, step=10000000, radius=radius, rings=1)
    
    return np.ndarray.flatten(fv)
    

# Test it
fv = image_patch_feature(char_dict["A"])
print(fv.shape)
```

So we get a 72-element feature vector for each character. This only works with sizes down to 10 pixels. Below that it refuses to produce a descriptor. Now we can make a list of features for each character in the set:

```python
feature_list = []
for cc in char_set:
    
    # Make a feature vector
    fv = image_patch_feature(char_dict[cc])
    
    # Add to list
    feature_list.append(fv)    
```

And now we can make a modified asciiise function to use this. We'll set it up to minimise a weighted sum of the brightness and DASIY feature distances.

```python
def asciiise2(image, char_set, brightness_list, feature_list, fscale, block_height, aspect_ratio):
    """
    Create ascii art string for an image
    """
    
    # Block size
    bh = int(block_height)
    bw = int(np.floor(block_height*aspect_ratio))
    
    # Number of blocks
    h,w = image.shape
    nc = int(np.floor(w/bw))
    nr = int(np.floor(h/bh))
    
    # Brightness grid
    brightness_grid = np.zeros((nr,nc))
    
    # Empty list to build our picture in
    asc = []
    
    # Loop through blocks
    for rr in range(nr):
        for cc in range(nc):
            
            # Grab an image patch
            patch = image[rr*bh:((rr+1)*bh-1), cc*bw:((cc+1)*bw-1)]
            
            # Get it's brightness
            brightness_grid[rr,cc] = image_patch_brightness(patch)
    
    # Rescale
    brightness_grid = (brightness_grid - np.min(brightness_grid) ) / (np.max(brightness_grid)-np.min(brightness_grid))
    brightness_list = (brightness_list - np.min(brightness_list) ) / (np.max(brightness_list)-np.min(brightness_list))
    
    # Loop through blocks
    for rr in range(nr):
        for cc in range(nc):
            
            # Block rescaled brightness
            bright = brightness_grid[rr,cc]
            
            # Block feature vector
            patch = image[rr*bh:((rr+1)*bh-1), cc*bw:((cc+1)*bw-1)]
            feature = image_patch_feature(patch)
            
            # Distances
            bright_dist = [np.abs(bright-cb) for cb in brightness_list]
            feat_dist = [np.linalg.norm(feature-cf) for cf in feature_list]
            
            # Weighted sum
            dist = list(np.array(bright_dist)+ fscale*np.array(feat_dist))
            
            # Find minimum distance
            idx = dist.index(min(dist))
            
            # Add character to the picture
            asc.append(char_set[idx])
            
        # New line
        asc.append("\n")
    
    # Print it
    print(''.join(asc))
```

Let's try that out.

```python
# Set the scale factor for the feature distance sum
fscale = 0.5

# And go!
asciiise2(test_image, char_set, brightness_list, feature_list, fscale, block_height, aspect_ratio)
```

Not really that much difference. I've spent a lot of time squinting at ascii Lenna with various values of fscale, and there's not a big subjective effect. However, I do think the feature matching improves things a bit. There's a lot more / and a lot less ! which makes the hat look less blocky. Ultimately, eyes are really most sensitive to brightness.
